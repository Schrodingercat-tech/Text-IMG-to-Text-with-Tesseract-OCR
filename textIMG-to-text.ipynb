```python
import pytesseract # creating an interface for tesseract-OCR 
import cv2 # for working with images
import matplotlib.pyplot as plt # for visualising text image
import os # for working with file paths
import pyttsx3 # for o/p our extracted text data into audio

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
path = "./OCR-text-imgs/"
images = []
for img in  os.listdir(path):
    images.append(
        cv2.imread(path+img)
        )

image = images[2]
imgchar = pytesseract.image_to_string(image)
plt.imshow(image)
print(imgchar)
```

```python
def speak(text):
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)  
    engine.setProperty('volume', 0.99) 
    engine.say(text)
    engine.runAndWait()

text_to_speak = imgchar
speak(text_to_speak)
```
